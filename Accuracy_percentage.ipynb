```python
# Convert dataframe to numpy arrary
data=data.to_numpy()
```

```python
# Print shape of numpy arrary
data.shape
```

```python
# Get training and testing dataset from numpy
training = data[0:600000]
testing = data[600000:]
```

```python
# Print training and testing dataset shape
print(training.shape)
print(testing.shape)
```

```python
# Seperate true labels from training and testing datasets
training_features = training[:,0:-1]
training_labels = training[:,-1]
testing_features = testing[:,0:-1]
testing_labels = testing[:,-1]
```

```python
# Print shape of input that training_feaures and true labels that is training_labels
print(training_features.shape)
print(training_labels.shape)
```

```python
# Checking input shape to neural network
training_features[0].shape
```

```python
# Import classes
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense, Dropout
```

```python
# Model definition
model = Sequential()
model.add(Input(shape=(6,)))
model.add(Dense(12,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(6,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(1,activation='sigmoid'))
```

```python
# Configure model training
model.compile(optimizer='sgd', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Train model
model.fit(training_features,training_labels,epochs=100,validation_data=(testing_features,testing_labels))
```
