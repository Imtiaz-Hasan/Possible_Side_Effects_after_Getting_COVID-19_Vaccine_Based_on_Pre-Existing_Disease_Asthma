```python
import pandas as pd
from matplotlib.pyplot import pie, axis, show
%matplotlib inline

df = pd.read_excel("/content/drive/MyDrive/dataset/main.xlsx")

sums = df.groupby(df["VAX_MANU"])["Sum"].sum()
axis('equal');
pie(sums, labels=sums.index, autopct='%1.1f%%');
show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
  
X = ['Moderna','Pfizer','Jansen']
Ygirls = [273924,242089,47395]
Zboys = [97337,103691,25533]
  
X_axis = np.arange(len(X))
  
plt.bar(X_axis - 0.2, Ygirls, 0.4, label = 'Female')
plt.bar(X_axis + 0.2, Zboys, 0.4, label = 'Male')
  
plt.xticks(X_axis, X)
plt.xlabel("Vaccine Name")
plt.title("Symptoms between Female vs Male")
plt.legend()

plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np

plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True

x = ['Moderna Female', 'Moderna Male', 'Pfizer Female', 'Pfizer Male', 'Janssen Female', 'Janssen Male']
y = [273924, 97337, 242089, 103691, 47395, 25533]

width = 0.35
fig, ax = plt.subplots()

pps = ax.bar(x, y, width, align='center')

for p in pps:
   height = p.get_height()
   ax.text(x=p.get_x() + p.get_width() / 2, y=height+.10,
      s="{}".format(height),
      ha='center')

plt.show()
```
