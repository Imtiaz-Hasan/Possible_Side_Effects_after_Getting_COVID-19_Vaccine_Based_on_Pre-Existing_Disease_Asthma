```python
#importing datasets 
df= pd.read_excel("/content/drive/MyDrive/Vaccine-history-dataset/new-dataset.xlsx")
```

```python
df.head()
```

```python
fdata=df[(df.HISTORY=="Asthma")&(df.VAX_MANU=="JANSSEN")&(df.SEX=="F") &(df['AGE_YRS']>=21) &(df['AGE_YRS']<=30)][["Total","VAX_MANU","AGE_YRS","SEX","HISTORY"]]
fdata

import matplotlib.pyplot as plt

from collections import Counter

print(Counter(fdata['Total']))
dict=Counter(fdata['Total'])
dict
```

```python
labels=[]
for x, y in dict.items():
  labels.append(str(x)+" types of symptoms")
sizes=[]
for x, y in dict.items():
  sizes.append(y)

plt.pie(sizes, labels=labels,
autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')
plt.title('Asthma,JANSSEN,Female (Age 21-30)')
plt.show()
```
