# **Symptoms affected on female**

```python
fdata=df[(df.SEX=="F") &(df['AGE_YRS']>=21) &(df['AGE_YRS']<=30)][["TOS","AGE_YRS"]]
fdata
```

```python
import matplotlib.pyplot as plt

from collections import Counter
```

```python
print(Counter(fdata['TOS']))
dict=Counter(fdata['TOS'])
dict
```

```python
labels=[]
for x, y in dict.items():
  labels.append(str(x)+" types of symptoms")
sizes=[]
for x, y in dict.items():
  sizes.append(y)
```

```python
plt.pie(sizes, labels=labels,
autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')
plt.title('Symptoms affected on Female (Age 21-30)')
plt.show()
```

```python
import pandas as pd
from matplotlib.pyplot import pie, axis, show
%matplotlib inline

df = pd.read_excel("/content/drive/MyDrive/dataset1/main2.xlsx")

sums = df.groupby(df["VAX_MANU"])["TOS"].sum()
axis('equal');
pie(sums, labels=sums.index, autopct='%1.1f%%');

show()
```

```python
from matplotlib import pyplot as plt
import numpy as np

plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True

x = ['Moderna Female', 'Moderna Male', 'Pfizer Female', 'Pfizer Male', 'Janssen Female', 'Janssen Male']
y = [273924, 97337, 242089, 103691, 47395, 25533]

width = 0.35
fig, ax = plt.subplots()

pps = ax.bar(x, y, width, align='center')

for p in pps:
   height = p.get_height()
   ax.text(x=p.get_x() + p.get_width() / 2, y=height+.10,
      s="{}".format(height),
      ha='center')

plt.show()
```
